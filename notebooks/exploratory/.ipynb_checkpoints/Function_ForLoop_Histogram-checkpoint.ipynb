Function for a for loop for making a histoplot:

```python
# Load libraries:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import KNNImputer
```

```python
# Read cleaned dataset

data_mmh = pd.read_csv('../../data/cleaned/data_mmh_cleaned1.csv')
data_mmh.head()
```

```python
#Function for a for loop for making a histoplot:
    
def plot_histograms(data, column_names):
    for column in column_names:
        plt.figure()  # Create a new figure for each plot
        sns.histplot(data[column], kde=True)
        plt.title(f'Histogram for {column}')
        plt.show()

# Example data
data = {
    'Age': [25, 30, 35, 28, 32, 31],
    'Income': [50000, 60000, 75000, 55000, 60000, 65000],
    'Score': [80, 85, 70, 90, 78, 88]
}

# Columns to plot
columns_to_plot = ['Age', 'Income', 'Score']

# Call the function to plot histograms for the selected columns
plot_histograms(data, columns_to_plot)
```

```python
#Function for a for loop for making a histplot:
    
def plot_histograms(data_mmh, column_names):
    for column in column_names:
        plt.figure()  # Create a new figure for each plot
        sns.histplot(data_mmh[column], kde=True)
        plt.title(f'Histogram for {column}')
        plt.show()

# Example data
data = data_mmh

# Columns to plot
#columns_to_plot = ['Age', 'Anxiety'] (numerical/categorical)
columns_to_plot = ['Age', 'Anxiety'] (numerical)

# Call the function to plot histograms for the selected columns
plot_histograms(data_mmh, columns_to_plot)
```

```python
Function for a for loop for making a histoplot with categorical and numerical columns:
    
def plot_categorical_histograms(data, categorical_column, numerical_column):
    categories = data[categorical_column].unique()
    
    for category in categories:
        subset = data[data[categorical_column] == category]
        plt.figure()
        sns.histplot(subset[numerical_column], kde=True)
        plt.title(f'Histogram for {numerical_column} in {categorical_column}={category}')
        plt.show()

# Example data
data = {
    'Category': ['A', 'A', 'B', 'A', 'B', 'B'],
    'Value': [10, 15, 20, 25, 30, 35]
}

# Call the function to plot histograms
plot_categorical_histograms(data, 'Category', 'Value')
```
